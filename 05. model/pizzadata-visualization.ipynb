# 2. Data Visualization

--------------------------------
**os.listdir(path)**는 path 경로에 있는 값들을 리스트화하는 함수

    classes -> ['not_pizza', 'pizza', 'food101_subset.py'] classes는 다음과 같은 값을 갖음

    os.listdir(path+'/'+classes[i]) 


**random.randint(a, b)**는 x를 출력하는데 이때 x는 a <= x <= b 이러한 범위를 갖고 x는 정수형이다.

    random.randint(0,1) -> 0 또는 1을 랜덤으로 출력
**cv2는 이미지를 다루는 모듈**로 imread로 데이터(이미지)를 읽고 cvtColor로 이미지의 색상을 변경한다.

--------------------------------

```python
path = '../input/pizza-not-pizza/pizza_not_pizza'
classes=os.listdir(path) 

plt.figure(figsize=(16,8))

for x in range(10):
    i = random.randint(0,1)
    images=os.listdir(path+'/'+classes[i]) 
    # -> i = 0이면 not_pizza 폴더에 있는 [이미지 이름들이 리스트화] 됨 i = 1이라면 pizza 폴더안에 있는...
    j=random.randint(0,900) # 각 폴더에는 983개의 이미지가 있어서 이와 같은 조건을 준 것 같음
    image = cv2.imread(path+'/'+classes[i]+'/'+images[j]) # 해당 경로에 있는 이미지 파일을 읽고 image 객체에 저장
    image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB) #cvtColor는 색변환 함수로 image의 BGR 색상 이미지를 RGB 색상 이미지로 변환
    ax = plt.subplot(5,5,x+1) #서브플롯을 5행 5열로 만들고 x+1에 해당하는 인덱스에 그림을 그림
    plt.imshow(image)
    plt.title(classes[i])
    plt.axis("off")   

plt.show()
```
